```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/media/prince/5A4E832F4E83034D/Visualization practice /Data/museum_visitors.csv')
```

```python
w = df['America Tropical Interpretive Center'].sum()
x = df['Firehouse Museum'].sum()
y = df['Avila Adobe'].sum()
z = df['Chinese American Museum'].sum()
labels = ['Firehouse Museum', 'Avila Adobe', 'Chinese American Museum', 'America Tropical Interpretive Center']
sizes = [w,x, y, z]
colors = ['magenta','blue', 'red', 'purple']
explode = (0.2, 0, 0,0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.title('Museum Visitors Distribution')
plt.show()
```
