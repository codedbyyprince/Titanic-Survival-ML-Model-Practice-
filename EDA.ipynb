```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6, 4))
plt.boxplot(
    df['Age'],
    vert=True,
    showmeans=True
)

plt.title('Boxplot of Age')
plt.ylabel('Age')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.boxplot(df['Fare'], showmeans=True)
plt.yscale('log')

plt.title('Boxplot of Fare (Log Scale)')
plt.ylabel('Fare (log)')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.boxplot(df['Fare'])

plt.ylim(0, 100)   # zoom where most data lives
plt.title('Fare Boxplot (Zoomed)')
plt.ylabel('Fare')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
plt.scatter(df['Age'], df['Fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')
plt.show()
```

```python
grouped_age = [
    df[df['Pclass'] == 1]['Age'],
    df[df['Pclass'] == 2]['Age'],
    df[df['Pclass'] == 3]['Age']
]

plt.boxplot(grouped_age, labels=['Class 1', 'Class 2', 'Class 3'], vert=True,
    showmeans=True
)
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.title('Age distribution by Passenger Class')
plt.show()
```

```python
plt.figure(figsize=(8, 5))

grouped = [
    df[df['Pclass'] == 1]['Fare'],
    df[df['Pclass'] == 2]['Fare'],
    df[df['Pclass'] == 3]['Fare']
]

plt.boxplot(
    grouped,
    labels=['Class 1', 'Class 2', 'Class 3'],
    vert=True,
    showmeans=True
)

plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.title('Fare distribution by Passenger Class')

plt.ylim(0, 300)
plt.yticks(range(0, 301, 10))

#  the magic you noticed
plt.grid(axis='y', linestyle='--', linewidth=0.9, alpha=0.8)

plt.tight_layout()
plt.show()
```

```python
counts = df['Pclass'].value_counts().sort_index()

plt.bar(counts.index, counts.values)
plt.xlabel("Passenger Class")
plt.ylabel("Number of People")
plt.title("Count of Passengers by Class")
plt.show()
```

```python
survival_rate = df.groupby('Pclass')['Survived'].mean()

plt.bar(survival_rate.index, survival_rate.values)
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.title("Effect of Passenger Class on Survival")
plt.ylim(0, 1)
plt.show()
```

```python
counts_sex = df['Sex'].value_counts().sort_index()
plt.bar(counts_sex.index, counts_sex.values)
plt.title('Sex distribution ')
plt.xlabel('male or female ')
plt.ylabel('Num in each gender')
```

```python
sex_rate = df.groupby('Sex')['Survived'].mean()
plt.bar(sex_rate.index , sex_rate.values)
plt.title('effect of gender on survival')
plt.xlabel('gender')
plt.ylabel('survival rate')
```

```python
grouped = [
    df[df['Survived'] == 0]['Age'],
    df[df['Survived'] == 1]['Age']
]

plt.boxplot(
    grouped,
    labels=['not survivied ', 'survived'],
    vert=True,
    showmeans=True
)

plt.xlabel('survived')
plt.ylabel('age')
plt.title('Affects of age on survival ')
plt.grid(axis='y', linestyle='--', linewidth=0.9, alpha=0.8)

plt.tight_layout()
plt.show()
```
