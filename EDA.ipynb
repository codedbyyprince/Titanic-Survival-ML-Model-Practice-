```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/media/prince/5A4E832F4E83034D/TItanic Project/cleaned.csv')
```

```python
df.info()
```

```python
pclass_counts = df['Pclass'].value_counts().sort_index()

plt.bar(pclass_counts.index, pclass_counts.values)
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")
plt.title("Distribution of Passenger Class")
plt.show()
```

```python
sex_counts = df['Sex'].value_counts().sort_index()

plt.bar(sex_counts.index, sex_counts.values)
plt.xlabel("Sex")
plt.ylabel("Number of Passengers")
plt.title("Distribution of Sex")
plt.show()
```

```python
plt.hist(df['Age'], bins=8, edgecolor='black')
plt.grid( axis='y', alpha=0.75)
```

```python
sibsp_counts = df['SibSp'].value_counts().sort_index()

plt.bar(sibsp_counts.index , sibsp_counts.values)
```

```python
parch_counts = df['Parch'].value_counts().sort_index()

plt.bar(parch_counts.index , parch_counts.values)
```

```python
plt.hist(df['Fare'], bins=30, edgecolor='black')
plt.grid( axis='y', alpha=0.75)
```

```python
embarked_counts = df['Embarked'].value_counts().sort_index()

plt.bar(embarked_counts.index , embarked_counts.values)
```

```python
plt.figure(figsize=(6, 4))
plt.boxplot(
    df['Age'],
    vert=True,
    showmeans=True
)

plt.title('Boxplot of Age')
plt.ylabel('Age')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.boxplot(df['Fare'], showmeans=True)
plt.yscale('log')

plt.title('Boxplot of Fare (Log Scale)')
plt.ylabel('Fare (log)')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.boxplot(df['Fare'])

plt.ylim(0, 100)   # zoom where most data lives
plt.title('Fare Boxplot (Zoomed)')
plt.ylabel('Fare')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
plt.scatter(df['Age'], df['Fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')
plt.show()
```

```python
grouped_age = [
    df[df['Pclass'] == 1]['Age'],
    df[df['Pclass'] == 2]['Age'],
    df[df['Pclass'] == 3]['Age']
]

plt.boxplot(grouped_age, labels=['Class 1', 'Class 2', 'Class 3'], vert=True,
    showmeans=True
)
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.title('Age distribution by Passenger Class')
plt.show()
```
